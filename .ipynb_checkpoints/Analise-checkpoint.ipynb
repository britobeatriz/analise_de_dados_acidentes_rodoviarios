# Análises

## Bibliotecas

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
```

## dataset

```python
dados = pd.read_csv('dados/dados.csv')
```

### Início das análises

```python
pd.set_option('display.max_columns', None)
dados.tail()
```

```python
dados['data_completa'] = pd.to_datetime(dados['data_completa'],format='%Y-%m-%d')
dados['ano'] = dados['data_completa'].dt.year.astype(str)
dados['ano'].unique()
```

Registros de acidentes ocorridos em rodovias federais no ano de 2010 a 2022.

### Ano em que mais ocorreu acidentes

```python
dados['ano'].value_counts().head()
```

2015 foi o ano com mais registros de acidentes.

### Dias da semana em que mais ocorre acidentes:

```python
# criando a fig e o ax no matplotlib
fig, ax = plt.subplots(figsize=(6,4))

# criando o gráfico no seaborn
sns.countplot(dados['dia_da_semana'], ax=ax, palette=['#2b5070','#2b5070','#2b5070','#2b5070','#0a2b47','#2b5070','#2b5070'])

# modificando o fundo
ax.set_frame_on(False)

# retirando o eixo Y
ax.get_yaxis().set_visible(False)

# retirando os ticks do eixo x
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')

# ajustando o titulo "dias da semana"
ax.set_xlabel('Dias da Semana', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')

# colocando os rótulos
for retangulo in ax.patches:
    ax.text(retangulo.get_x() + retangulo.get_width() / 2,
           retangulo.get_height() + 22,
           '{:,}'.format(int(retangulo.get_height())).replace(',','.'),
           ha='center',
           fontsize=11,
           color='grey')
```

Sexta-feira é o dia que mais ocorre acidentes, seguido de sábado e domingo. Sendo terça-feira o dia que menos ocorre acidentes. 

### Dia da semana que mais ocorre acidentes com pessoas gravemente feridas e mortas

```python
df_g_feridos = dados.groupby(['dia_da_semana']).gravemente_feridos.sum().sort_values(ascending=False).head(7)
df_mortos = dados.groupby(['dia_da_semana']).mortos.sum().sort_values(ascending=False).head(7)
```

```python
fig,ax = plt.subplots(figsize=(6,4))
plt.title("Dia com mais frequência de gravemente feridos",fontsize=15)
plt.ylabel('Número de feridos',fontsize=15,color='#4c4c4c')
plot = df_g_feridos.plot.bar(color='#0a2b47')
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')
ax.set_xlabel('Dia da semana', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')
plt.tight_layout();

fig,ax = plt.subplots(figsize=(6,4))
plt.title("Dia com mais frequência de mortos",fontsize=15)
plt.ylabel('Número de feridos',fontsize=15,color='#4c4c4c')
plot = df_mortos.plot.bar(color='#2b5070')
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')
ax.set_xlabel('Dia da semana', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')
plt.tight_layout();
```

Apesar de Sexta-feira ser o dia que possui mais registros de acidentes, Domingo e Sábado, são os dias em que há maior ocorrencia de acidentes graves e com mortos.  

### Fase do dia em que mais ocorre acidentes:

```python
# criando a fig e o ax no matplotlib
fig, ax = plt.subplots(figsize=(6,4))

# criando o gráfico no seaborn
sns.countplot(dados['fase_do_dia'], ax=ax, palette=['#2b5070','#0a2b47','#2b5070'])

# modificando o fundo
ax.set_frame_on(False)

# retirando o eixo Y
ax.get_yaxis().set_visible(False)

# retirando os ticks do eixo x
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')

# ajustando o titulo "fase do dia"
ax.set_xlabel('Fase do Dia', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')

# colocando os rótulos
for retangulo in ax.patches:
    ax.text(retangulo.get_x() + retangulo.get_width() / 2,
           retangulo.get_height() + 22,
           '{:,}'.format(int(retangulo.get_height())).replace(',','.'),
           ha='center',
           fontsize=11,
           color='grey')
```

O gráfico mostra que a maioria dos acidentes acontecem a tarde.

obs: considerando o período da manhã das 06:00:00 as 11:59:59; tarde, 12:00:00 as 17:59:59; noite, 18:00:00 as 23:59:59; madrugada, 00:00:00 as 05:59:59.

### Horário em que acidentes são mais frequentes:

```python
# criando DataFrame com apenas 10
df_horario = dados['horario'].value_counts().head(10)

# criando a fig e o ax no matplotlib
fig, ax = plt.subplots(figsize=(6,4))

# plotando gráfico no matplotlib
df_horario.plot.bar(color='#2b5070')

# modificando o fundo
ax.set_frame_on(False)

# retirando o eixo Y
ax.get_yaxis().set_visible(False)

# retirando os ticks do eixo x
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')

# ajustando o titulo "Horários com mais frequência de acidente"
ax.set_xlabel('Horários com mais frequência de acidente', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')

# colocando os rótulos
for retangulo in ax.patches:
    ax.text(retangulo.get_x() + retangulo.get_width() / 2,
           retangulo.get_height() + 22,
           '{:,}'.format(int(retangulo.get_height())).replace(',','.'),
           ha='center',
           fontsize=11,
           color='grey')
plt.tight_layout();
```

Apesar do gráfico anterior mostrar que a maioria dos acidentes acontece no período da tarde, esse mostra que o horário com mais frequência de acidentes é por volta das 18:00:00 horas.

### Rodovias com mais registros de acidentes

```python
dados['trecho'].value_counts().head()
```

A rodovia com maior registro de acidentes é a BR-101/SC.

### Tipos de acidentes mais frequentes

```python
dados_tipo_acidente = dados['tipo_de_acidente'].value_counts().head(10)

# criando a fig e o ax no matplotlib
fig, ax = plt.subplots(figsize=(6,4))

# modificando o fundo
ax.set_frame_on(False)

# retirando os ticks do eixo x
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')

# ajustando o titulo "tipo de acidente"
ax.set_xlabel('Tipo de acidente', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')

# plotando gráfico no matplotlib
dados_tipo_acidente.plot.bar(color='#2b5070')
plt.show()
```

Colisão traseira é o tipo mais comum de acidente.

### Tipos de acidentes com mais vitimas fatais

```python
df_v_f = dados.groupby(['tipo_de_acidente']).mortos.sum().sort_values(ascending=False).head(10)

# criando a fig e o ax no matplotlib
fig, ax = plt.subplots(figsize=(6,4))

# modificando o fundo
ax.set_frame_on(False)

# retirando os ticks do eixo x
ax.tick_params(axis='x', length=0, labelsize=12, colors='grey')

# ajustando o titulo "tipo de acidente"
ax.set_xlabel('Tipo de acidente', labelpad=20, fontdict={'fontsize':14}, color='#4c4c4c', loc='center')

# plotando gráfico no matplotlib
df_v_f.plot.bar(color='#2b5070')
plt.show()
```

Colisão frontal é o tipo de acidente com mais vítimas fatais.

# Conclusões

- **2015** foi o ano com mais registro de acidentes;
- **Sexta-feira** é o dia que mais ocorre acidentes;
- A maior parte dos acidentes ocorrem a **tarde**;
- **18h** é o horario que mais ocorre acidentes;
- A rodovia com maior registro de acidentes é a **BR-101/SC**;
- **Colisão Traseira** é o tipo de acidente mais comum;
- **Sábado** e **Domingo**, são os dias com maior ocorrencia de acidentes com pessoas gravemente feridas e mortas;
- **Colisão frontal** é o tipo de acidente com mais vítimas fatais.
